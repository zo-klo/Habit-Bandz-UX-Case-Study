# Habit Bandz User Insight Workflow

## Project Overview

This notebook presents a cleaned, portfolio-friendly version of the Habit Bandz insight workflow using synthetic data. The goal is to demonstrate how qualitative participant feedback can be translated into structured themes, lightweight text analysis, and product direction without exposing private research materials.

The workflow combines:

- direct review of participant-style feedback
- structured cleaning and preprocessing
- thematic synthesis of friction points and value signals
- lightweight text analysis to support pattern detection
- product interpretation grounded in behavioral context


```python
from pathlib import Path

import pandas as pd
from IPython.display import display

from habitbandz_portfolio_utils import make_frequency_table, prepare_feedback_dataframe, top_words

pd.set_option("display.max_colwidth", 120)
pd.set_option("display.max_columns", None)

DATA_PATH = Path("data/habitbandz_synthetic_feedback.csv")
```

## Research Objective

The objective of this workflow is to move from open-ended user feedback to decision-ready product insight. In particular, the analysis asks:

- What recurring patterns appear in descriptions of urges, triggers, and barriers to change?
- Which pain points cluster around intervention speed, logging burden, and progress interpretation?
- How can these patterns be synthesized into product recommendations for a behavior-change intervention?


## Data Overview

The public workflow uses a small synthetic dataset with the same general structure as the original project: habit description, triggers, challenges, prior interventions, liked features, and product improvement requests.


```python
raw_df = pd.read_csv(DATA_PATH)

print(f"Rows: {raw_df.shape[0]}")
print(f"Columns: {raw_df.shape[1]}")

display(raw_df.head())

overview = pd.DataFrame(
    {
        "column": raw_df.columns,
        "non_null_count": raw_df.notna().sum().values,
        "missing_rate": raw_df.isna().mean().round(2).values,
    }
)

display(overview)
```

## Cleaning and Preprocessing

The preprocessing layer is intentionally lightweight. Each participant response is consolidated into a single analysis field and then normalized through lowercasing, punctuation removal, and whitespace cleanup. This keeps the workflow interpretable while still supporting frequency-based review.


```python
df = prepare_feedback_dataframe(raw_df)

display(
    df[
        [
            "habit_desc",
            "triggers",
            "all_text",
            "all_text_clean",
        ]
    ].head(3)
)
```

## Manual Qualitative Review

For a small-N UX project, direct reading still matters. Instead of printing every response as raw text, the notebook keeps the review compact and structured so it is easier to compare patterns across participants.


```python
review_columns = [
    "habit_desc",
    "challenges",
    "triggers",
    "liked_features",
    "app_improve",
]

display(df[review_columns])
```

## Thematic Synthesis

The synthesis below captures the highest-signal themes from the synthetic sample. These themes are framed as product-relevant patterns rather than generic observations.


```python
theme_summary = pd.DataFrame(
    [
        {
            "theme": "Awareness and Automaticity",
            "evidence": "Participants describe noticing the behavior late or only after it is already in progress.",
            "product_meaning": "The intervention has to work before reflection is available; low-friction interruption matters more than rich input.",
        },
        {
            "theme": "Stress-Responsive Use Cases",
            "evidence": "Stress, fatigue, meetings, and transitions repeatedly appear as triggers.",
            "product_meaning": "Usability must hold up in low-attention and high-cognitive-load contexts, not just calm review sessions.",
        },
        {
            "theme": "Value of Physical Intervention",
            "evidence": "The wearable cue is consistently described as helpful when it interrupts the behavior quickly.",
            "product_meaning": "Tactile interruption appears to be part of the core product value proposition rather than a secondary feature.",
        },
        {
            "theme": "Need for Lower Logging Burden",
            "evidence": "Participants ask for fewer steps, simpler navigation, and one-handed use.",
            "product_meaning": "Logging friction is likely to suppress usage during exactly the moments the product is meant to support.",
        },
        {
            "theme": "Progress Interpretation",
            "evidence": "Several users want clearer trends and more confidence that progress is happening over time.",
            "product_meaning": "Reflection features should not only store data; they should help users and clinicians interpret change.",
        },
    ]
)

display(theme_summary)
```

## Lightweight Text Analysis

This step uses basic frequency review to support, not replace, close reading. For this kind of dataset, the purpose is to reinforce pattern detection and reduce blind spots rather than to produce a standalone model.


```python
overall_terms = make_frequency_table(df["all_text_clean"], min_df=2, n=12)
trigger_terms = top_words(df["triggers"], min_df=1, n=8)
improvement_terms = top_words(df["app_improve"], min_df=1, n=8)

print("Overall repeated terms")
display(overall_terms)

print("Trigger-related terms")
display(trigger_terms)

print("Improvement-related terms")
display(improvement_terms)
```

## Key Product Insights

The strongest story in this workflow is not the text analysis alone. It is the translation of mixed-method signals into product direction.


```python
product_insights = pd.DataFrame(
    [
        {
            "insight": "Optimize for intervention speed",
            "why_it_matters": "Automatic and stress-reactive behaviors leave little room for multi-step interaction.",
            "product_direction": "Prioritize fast, low-attention interruption flows and reliable confirmation feedback.",
        },
        {
            "insight": "Treat logging friction as a product risk",
            "why_it_matters": "If capture is too slow or cognitively heavy, users may skip the product when it matters most.",
            "product_direction": "Simplify entry points, reduce taps, and design explicitly for one-handed and in-motion use.",
        },
        {
            "insight": "Separate intervention from reflection needs",
            "why_it_matters": "Some users need immediate interruption, while others also care deeply about trend interpretation.",
            "product_direction": "Design distinct experiences for in-the-moment support versus longitudinal progress review.",
        },
        {
            "insight": "Build toward personalized pathways",
            "why_it_matters": "Users differ in awareness, urgency, and tolerance for cognitive load.",
            "product_direction": "Use behavioral segmentation to inform default workflows, prompts, and follow-up support.",
        },
    ]
)

display(product_insights)
```

## Limitations

- The dataset in this notebook is synthetic and is intended only to demonstrate the workflow structure.
- Lightweight frequency review can support interpretation, but it should not be mistaken for a comprehensive language model or clinical analysis.
- A small sample is appropriate for early product direction, but not for strong behavioral claims.
- The output of this notebook should be treated as a decision-support layer, not as final validation.
